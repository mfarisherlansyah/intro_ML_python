```python
# LDA

import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import FactorAnalysis
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

iris = datasets.load_iris()

X = iris.data
y = iris.target
target_names = iris.target_names

# Transformasi dengan PCA
pca = PCA(n_components = 2)
X_r = pca.fit(X).transform(X)

# LDA itu supervised, membutuhkan label kelas
# Transformasi dengan LDA
lda = LinearDiscriminantAnalysis(n_components = 2)
X_r2 = lda.fit(X, y).transform(X)

# Transformasi dengan FA
fa = FactorAnalysis(n_components = 2)
X_r3 = fa.fit_transform(iris.data)

# Berapa persen PCA?
print("Explained variance ratio (first two components): %s" % str(pca.explained_variance_ratio_))


plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
# Plot fitur aslinya

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X[y == i, 0], X[y == i, 1], color = color, alpha = 0.8, lw = lw, label = target_name)
plt.legend(loc = 'best', shadow = False, scatterpoints = 1)
plt.title('Fitur aslinya dari IRIS dataset')


# Plot hasil PCA
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r[y == i, 0], X_r[y == i, 1], color = color, alpha = 0.8, lw = lw, label = target_name)
plt.legend(loc = 'best', shadow = False, scatterpoints = 1)
plt.title('PCA of IRIS dataset')

# Berapa persen LDA?
print("Explained variance ratio (first two components): %s" % str(lda.explained_variance_ratio_))

# Plot hasil LDA
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r2[y == i, 0], X_r2[y == i, 1], color = color, alpha = 0.8, lw = lw, label = target_name)
plt.legend(loc = 'best', shadow = False, scatterpoints = 1)
plt.title('LDA of IRIS dataset')

# Plot hasil FA
plt.figure()
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_r3[y == i, 0], X_r3[y == i, 1], color = color, alpha = 0.8, lw = lw, label = target_name)
plt.legend(loc = 'best', shadow = False, scatterpoints = 1)
plt.title('FA of IRIS dataset')

plt.show()


# Original
print("Original 2 Features")
(trainx, testx, trainy, testy) = train_test_split(X[:, 0:2], y, random_state = 3, test_size = 0.25)
model = MLPClassifier()
model.fit(trainx, trainy)
predictions = model.predict(testx)
print(classification_report(testy, predictions, target_names = target_names))

print("PCA")
(trainx, testx, trainy, testy) = train_test_split(X_r, y, random_state = 3, test_size = 0.25)
model = MLPClassifier()
model.fit(trainx, trainy)
predictions = model.predict(testx)
print(classification_report(testy, predictions, target_names = target_names))

print("LDA")
(trainx, testx, trainy, testy) = train_test_split(X_r2, y, random_state = 3, test_size = 0.25)
model = MLPClassifier()
model.fit(trainx, trainy)
predictions = model.predict(testx)
print(classification_report(testy, predictions, target_names = target_names))

print("FA")
(trainx, testx, trainy, testy) = train_test_split(X_r3, y, random_state = 3, test_size = 0.25)
model = MLPClassifier()
model.fit(trainx, trainy)
predictions = model.predict(testx)
print(classification_report(testy, predictions, target_names = target_names))

print("Original All Features")
(trainx, testx, trainy, testy) = train_test_split(X, y, random_state = 3, test_size = 0.25)
model = MLPClassifier()
model.fit(trainx, trainy)
predictions = model.predict(testx)
print(classification_report(testy, predictions, target_names = target_names))
```
